```python
import numpy as np 
import pandas as pd
df=pd.read_csv("./chipotle.tsv",sep="\t")
df
```

```python
df_copy=df.copy()
df_copy
df.isnull().sum()
df.dtypes
```

```python
df["item_price"]=df["item_price"].str.replace("$","")

df["item_price"]=df["item_price"].astype(float)
df
```

```python
df.dtypes

df["choice_description"]=df["choice_description"].str.replace("[","")
df["choice_description"]=df["choice_description"].str.replace("]","")
#df["choice_description"]=df["choice_description"].str.replace(np.nan,"NONE")
df
```

```python
# Calculate total quantity sold for each item
item_quantity = df.groupby('item_name')['quantity'].sum()
# print(item_quantity)
# Sort the items based on total quantity in descending order and fetch the top 5
top_5_items = item_quantity.sort_values(ascending=False).head(5)

# Display the result
print("Top 5 items based on total quantity sold:")
print(top_5_items)
```

```python
# Calculate total quantity sold for each item
choice_quantity = df.groupby("choice_description")['quantity'].sum()
# print(item_quantity)
# Sort the items based on total quantity in descending order and fetch the top 5
top_5_choice = choice_quantity.sort_values(ascending=False).head(5)

# Display the result
print("Top 5 choices based on total quantity sold:")
print(top_5_choice)
```

```python
# Calculate total quantity sold for each item
item_quantity = df.groupby('item_name')['quantity'].sum()
total_order = df.groupby('order_id')['quantity'].count()

print(total_order)
```

```python
revenue_per_product=df.groupby(['item_name', 'choice_description'])['item_price'].sum()
print(revenue_per_product)
```

```python
#Whta is average revenue amount per order?
revenue_per_order=df.groupby(['order_id'])['item_price'].mean()
print(revenue_per_order)
```
