```python
import numpy as np 
import pandas as pd
data_frame=pd.read_csv("./Data-cleaning-for-beginners-using-pandas.csv")

copy_df=data_frame
print(copy_df)
```

```python
data_frame.isnull().sum()
averageage = data_frame["Age"].mean()
averagerating=data_frame["Rating"].mean()
averageage
averagerating
data_frame["Age"] = data_frame.Age.fillna(averageage)
data_frame["Rating"] = data_frame.Rating.fillna(averagerating)

data_frame["Rating"]
```

```python
data_frame["Age"] = data_frame.Age.round(decimals=1)

data_frame["Rating"] = data_frame.Rating.round(decimals=1)
data_frame.tail(15)
data_frame["Salary"]=data_frame["Salary"].str.replace('k','000')
data_frame["Salary"]=data_frame["Salary"].str.replace('$','')
data_frame.tail(15)
```

```python
data_frame["Location"]=data_frame["Location"].str.replace(' Aus','')
data_frame["Location"]=data_frame["Location"].str.replace(',In','')
data_frame["Location"]=data_frame["Location"].str.replace(',Ny','')
data_frame["Location"]=data_frame["Location"].str.replace(' In','')
```

```python
data_frame["Easy Apply"]=data_frame["Easy Apply"].str.replace('-1','FALSE')
data_frame["Established"]=data_frame["Established"].replace(-1,np.nan)
averageestablished=data_frame["Established"].mean()
data_frame["Established"] = data_frame.Established.fillna(averageestablished)
data_frame["Established"] = data_frame.Established.round(decimals=0)
data_frame['Established'] = data_frame['Established'].astype(int)
data_frame.head(15)
```

```python
data_frame["Rating"]=data_frame["Rating"].replace(-1.0,0)
data_frame.tail(15)
```
